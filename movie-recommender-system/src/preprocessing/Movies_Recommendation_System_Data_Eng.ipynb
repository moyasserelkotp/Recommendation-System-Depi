## 🎯 Problem Statement (MovieLens Dataset)

#####  In the modern digital entertainment landscape, users face a constant challenge: what to watch next? With 'thousands of movie options available on various streaming platforms, making a decision becomes difficult without personalized guidance. Generic `Top 10` lists or trending movies often fail to reflect the unique tastes of individual users.

##### To solve this, we aim to build an **intelligent movie recommendation system** that uses the MovieLens dataset—a widely-used benchmark dataset containing millions of movie ratings provided by real users. The MovieLens dataset offers a rich foundation to analyze user preferences, movie metadata (titles, genres, release dates), and rating behaviors.

##### The key challenge lies in extracting valuable insights from this large and sparse user-item interaction matrix, and then transforming those insights into meaningful, personalized suggestions. A system that can learn from both user behavior patterns and movie features is critical to improving user satisfaction and engagement.

##### The goal is to build a solution that:
  - Understands user preferences through historical ratings.
  - Explores the content side (genre, title, etc.) to recommend similar movies.
  - Uses hybrid recommendation models to combine collaborative filtering and content-based filtering.
  - Can be deployed as an interactive application with real-time suggestions and updates.

### 2.2.3 Temporal Data Conversion

```python
# Calculate rating count and average rating per user
user_stats = ratings.groupby("userId").agg(
    rating_count=("rating", "count"),
    rating_avg=("rating", "mean")
).reset_index()
```

```python
# Calculate average rating per movie
movie_stats = ratings.groupby("movieId").agg(
    ave_rating=("rating", "mean")
).reset_index()
```

```python
# Extract year from the movie title
movies["year"] = movies["title"].str.extract(r"\((\d{4})\)")
```

```python
# Convert genres into separate columns
genres_expanded = movies["genres"].str.get_dummies(sep="|")
movies = pd.concat([movies, genres_expanded], axis=1)
```

```python
# Merge movie statistics with the movies dataset
movies = movies.merge(movie_stats, on="movieId", how="left")
```

```python
# Merge user statistics with the ratings dataset
ratings = ratings.merge(user_stats, on="userId", how="left")
```

```python
ratings["timestamp"] = pd.to_datetime(ratings['timestamp'], unit='s')
```
